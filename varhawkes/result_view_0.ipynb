```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import seaborn as sns
import numpy as np
import json

#userlist = ['s002','s032','s036','s047','s052']
user = 's032'
#start_line = 0
#end_line = 50
start_line_list = [0,50,100,150,200,250,300,350]
decay_rate = 0.005
#decay_rate_list = [0.001,0.01,0.1,1.0,2.0,5.0,10]
filename_list = ['0']*len(start_line_list)
path = './result/compare2/'


def filename(user,start_line,end_line,decay_rate):
    return user + '_'+str(start_line) + '_' + str(end_line) + 'decay'+str(decay_rate)+'.json'

for index,start_line in enumerate(start_line_list):
    end_line = start_line + 50
    filename_list[index] = path + filename(user,start_line,end_line,decay_rate)

  
   
#
for param_filename in filename_list:
    with open(param_filename,'r') as param_file:
        param_dict = json.load(param_file)        
        
    adjacency = np.array(param_dict['vi_exp']['adjacency']).reshape(11,11)
    #print(adjacency)
    mu = np.array(param_dict['vi_exp']['mu']).tolist()
    #print(mu.tolist())
    #print(mu)

   
    cmap = sns.cubehelix_palette(start = 1, rot = 3, gamma=0.8, as_cmap = True)
    sns.heatmap(adjacency, linewidths = 0.05, vmax=10, vmin=0, cmap=cmap, center=None, robust=False, annot=False)
   
    plt.show()
```
