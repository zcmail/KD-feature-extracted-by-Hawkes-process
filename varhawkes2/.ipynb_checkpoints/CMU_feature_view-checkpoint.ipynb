```python
import pandas as pd
import matplotlib.pyplot as plt

fileName = './CMU_oral_output.xls'
df_oral = pd.read_excel(fileName)
```

```python
ud_features = [column for column in df_oral.columns if 'UD' in column]
dd_features = [column for column in df_oral.columns if 'DD' in column]
hold_features = [column for column in df_oral.columns if 'H' in column]
```

```python
#user_list = ['s002','s032','s036','s047','s052']
#object_user = 's002'
#df = df_oral[df_oral['subject'] == object_user]

X = df_oral.drop(['subject','rep','sessionIndex'], axis=1)[dd_features]
xticks = [i*100 for i in range(20)] 

for i,feature in enumerate(dd_features): 
    X_slice = X[feature]
    plt.figure(figsize=(20, 4))
    plt.plot(X_slice)
    plt.title('raw features of '+feature)
    plt.xlabel('sample')
    plt.ylabel('interval')
    plt.xticks(xticks, size=14)
    plt.grid()
    plt.savefig('raw_features_of_'+feature+'.png')
```

```python
X = df_oral.drop(['subject','rep','sessionIndex'], axis=1)[hold_features]
xticks = [i*100 for i in range(20)] 

for i,feature in enumerate(hold_features): 
    X_slice = X[feature]
    plt.figure(figsize=(20, 4))
    plt.plot(X_slice)
    plt.title('raw features of '+feature)
    plt.xlabel('sample')
    plt.ylabel('interval')
    plt.xticks(xticks, size=14)
    plt.grid()
    plt.savefig('raw_features_of_'+feature+'.png')
```
